```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Olympic Games medal Dataset(from 1896 to 2018).csv"
gm_df = pd.read_csv(csv_file)
gm_df.head()
```

```python
#rename column
gm_df = gm_df.rename(columns={"Team(IOC Code)":"Country"})
gm_df.head()
```

```python
#split country and code
column_gm_df = gm_df['Country'].str.rsplit("(", expand=True)
country_column = column_gm_df[0]

#split codes again
column_gm_df = column_gm_df[1].str.split(expand=True)

#drop extraneous columns
column_gm_df = column_gm_df.drop(columns=[1, 2, 3, 4, 5])

country_code = column_gm_df[0]
```

```python
#combine with original df
#reorder columns
gm_df['Code'] = country_code
gm_df['Country'] = country_column
gm_df['Code'] = gm_df.Code.str.strip('()')
gm_df = gm_df[['Country', 'Code', 'summer_rank', 's_gold', 's_silver', 's_bronze',
       'summer_total', 'winter_rank', 'w_gold', 'w_silver', 'w_bronze',
       'winter_total', 'overall_rank', 'total_gold', 'total_silver',
       'total_bronze', 'total_medals']]
```

```python
#strip columns with comma'd numbers of commas
#convert those columns to int
gm_df['s_gold'] = gm_df['s_gold'].str.replace(",","")
gm_df['summer_total'] = gm_df['summer_total'].str.replace(",","")
gm_df['total_gold'] = gm_df['total_gold'].str.replace(",","")
gm_df['total_medals'] = gm_df['total_medals'].str.replace(",","")

gm_df = gm_df.astype({'s_gold':'int64', 'summer_total':'int64', 'total_gold':'int64', 'total_medals':'int64'})

gm_df.head(25)
```

```python
gm_df.to_csv('cleaned_gm.csv', index=False)
```
