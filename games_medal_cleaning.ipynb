```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Olympic Games medal Dataset(from 1896 to 2018).csv"
gm_df = pd.read_csv(csv_file)
gm_df.head()
```

```python
#split the country name and ioc code
gm_df = gm_df.rename(columns={"Team(IOC Code)":"Country"})
gm_df.head()
```

```python
column_gm_df = gm_df['Country'].str.split(expand=True)

column_gm_df = column_gm_df.drop(columns=[2, 3, 4, 5, 6, 7])
column_gm_df = column_gm_df.rename(columns={0:"Country", 1:"Code"})

column_gm_df.head()
```

```python
gm_df['Code'] = column_gm_df['Code']
gm_df['Country'] = column_gm_df['Country']
gm_df['Code'] = gm_df.Code.str.strip('()')
gm_df = gm_df[['Country', 'Code', 'summer_rank', 's_gold', 's_silver', 's_bronze',
       'summer_total', 'winter_rank', 'w_gold', 'w_silver', 'w_bronze',
       'winter_total', 'overall_rank', 'total_gold', 'total_silver',
       'total_bronze', 'total_medals']]
gm_df.head()
```

```python
gm_df.to_csv('cleaned_gm.csv')
```
