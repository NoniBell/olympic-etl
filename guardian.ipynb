```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# read summer data
summer_guardian_df = pd.read_csv('Guardian_files/Resources/summer.csv')

# add 'Season' column for summer data
summer_guardian_df['Season'] = 'Summer'
summer_guardian_df.head()
```

```python
# read winter data
winter_guardian_df = pd.read_csv('Guardian_files/Resources/winter.csv')

# add 'Season' column for winter data
winter_guardian_df['Season'] = 'Winter'
winter_guardian_df.head()
```

```python
# append winter data to summer data in a new dataframe
guardian_df = summer_guardian_df.append(winter_guardian_df)
guardian_df.head()
```

```python
# sort rows by 'Year'
guardian_df.sort_values('Year')
```

```python
# lowercase athlete name
guardian_df['Athlete'] = guardian_df['Athlete'].apply(str.lower)
guardian_df
```

```python
# save cleaned dataframe as csv
guardian_df.to_csv('Guardian_files/Resources/athlete_medals_cleaned.csv',index=False)
```

```python
# create a database connection
connection_string = "postgres:postgres@localhost:5432/Olympics_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# confirm table names
engine.table_names()
```

```python
# load dataframe into database
guardian_df.to_sql(name='athlete_medals', con=engine, if_exists='append', index=False)
```
